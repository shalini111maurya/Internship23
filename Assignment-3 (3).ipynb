```python
url='https://www.amazon.com/'
driver.get(url)
time.sleep(5)

key=input("Enter product to search:")

# Entering guitar in search field 
search_product=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[2]/div[1]/input').send_keys(key)
search_product

#  clucking in search field 
search_button=driver.find_element(By.XPATH,'/html/body/div[1]/header/div/div[1]/div[2]/div/form/div[3]/div/span/input')
search_button.click()
```

```python
# creating empty lists for scraping data

product_name=[]
price=[]
return_exchange=[]
exp_delivery=[]
product_url=[]

time.sleep(3)
#scrapping the required details
start=0
end=2

#scraping product name
for page in range(start,end):#for loop for scrapping 3 page
    url=driver.find_elements(By.XPATH,"//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
    for i in url: # getting url for every product
        product_url.append(i.get_attribute('href'))
    nxt_button=driver.find_elements(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")#scraping the list of buttons from the page
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
len(product_url)
```

```python
#Now scraping each products details

#Scraping product titles:
for ur in product_url:
    driver.get(ur)
    name=driver.find_elements(By.XPATH,'//h1[@class="a-size-large a-spacing-none"]')
    for i in name:
        if i.text is None:
            product_name.append("--")
        else:
            product_name.append(i.text) # Apending product title in the list

len(product_name)
```

```python
price=[]
for ur in product_url:
    driver.get(ur)
    prices=driver.find_elements(By.XPATH,'//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
    for i in prices:
        if i.text is None:
            price.append("Currently Unavailable")
        else:
            price.append(i.text)

len(price)
```

```python
url='https://www.google.com/imghp?hl=en'
driver.get(url)
time.sleep(5)
list1=["fruits","cars","Machine Learning","guitar","cakes"]
source=[]

for i in list1:
    search_product=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(i)
    search_product
    search_button=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
    search_button.click()
    time.sleep(5)
    img_urls=[]
    image=driver.find_elements(By.XPATH,'//div[@class="wXeWr islib nfEiy"]')
    for i in image:
        source=i.get_attribute('src')
        if source is not None:
            if(source[0:10]=='http'):
                img_urls.append(source)
                
                
    for  i in range(len(img_urls)):
        if i>10:
            break
    print("Downloading {0} of {1} images".format(i,10))
    response=requests.get(img_urls[i])
    file=open(r"C:\Users\Shalini Maurya\Downloads\img"+str(i)+".jpg","wb")
    file.write(response.content)
```
