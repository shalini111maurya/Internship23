```python
#Activating the chrome browser
ser=Service(r"C:\Users\Shalini Maurya\Downloads\chromedriver_win32\chromedriver.exe")
driver=webdriver.Chrome(service=ser)
```

```python
url ="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(5)

#extract rank
rank=[]
r=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[1][@align="center"]')
for i in r:
    rank.append(i.text)

#name
name=[]
n=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[2]')
for i in n:
    name.append(i.text)
    
#artist
artist=[]
a=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[2]')
for i in a:
    artist.append(i.text)

#upload date
upload_date=[]
ud=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[3]')
for i in ud:
    upload_date.append(i.text)

#view
view=[]
v=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"][1]//tbody//tr//td[4]')
for i in v:
    view.append(i.text)

dt=pd.DataFrame({"Rank":rank, "Name":name,"Artist":artist,"Upload_Date":upload_date,"View":view})
dt

```

# Q9.Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(5)

#extracting Name:
name=[]
n=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]//a')
for i in n:
    name.append(i.text)

#ectracting Year Span
year=[]
ys=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
for i in ys:
    year.append(i.text)
    
#extracting Genre
genre=[]
genr = driver.find_elements_by_xpath('//span[@class="genre"]')
for i in genr:
    genre.append(i.text)
    
    
#extracting run time:
run=[]
r=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
for i in r:
    run.append(i.text)

#extracting ratings
rating=[]
rt=driver.find_elements(By.XPATH,'//div[@class="ipl-rating-star small"]//span[2]')
for i in rt:
    rating.append(i.text)
    
    
#extracting Votes
votes=[]
vt=driver.find_elements(By.XPATH,'//span[@name="nv"]')
for i in vt:
    votes.append(i.text)

#making DataFrame
dt=pd.DataFrame({"Name":name,"Year Span":year,"Genre":genre,"Run Time":run,"Rating":rating,"votes":votes})
dt
```

# 10. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the ShowAllDataset page through code

```python
url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(5)

#Click on View all Datasets
driver.find_element(By.XPATH,'//td//span[2][@class="whitetext"]//a').click()

#scraping Dataset Name
dtn=[]
dt=driver.find_elements(By.XPATH,'//p[@class="normal"]//b//a')
for i in dt:
    dtn.append(i.text)

#sraping Task
task=[]
tk=driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[3]//p[@class="normal"]')
for i in tk:
    task.append(i.text)
    
#sraping Attribute Type
attribute=[]
at=driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[4]//p[@class="normal"]')
for i in at:
    try:
        attribute.append(i.text)
    except NOSuchElementFound:
        attribute.append("--")

#scraping Number of Instances
nins=[]
ni=driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[5]//p[@class="normal"]')
for i in ni:
    try:
        nins.append(i.text)
    except NOSuchElementFound:
        nins.append("--")
        
#scraping Number of Attribute
natb=[]
atb=driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[6]//p[@class="normal"]')
for i in atb:
    try:
        natb.append(i.text)
    except NOSuchElementFound:
        natb.append("--")
    
#scraping Year
year=[]
y=driver.find_elements(By.XPATH,'//table[2]//tbody[1]//tr[1]//td[7]//p[@class="normal"]')
for i in y:
    try:
        year.append(i.text)
    except NOSuchElementFound:
        year.append("--")
```

```python
dt=pd.DataFrame({"Dataset Name":dtn,"Task":task,"Attribute Type":attribute,"Number of Instances":nins,"Number of Attribute":natb,"Year":year})
dt.head(100)
```
