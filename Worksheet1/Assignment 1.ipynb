```python
# A Program to find a Factorial
num= int(input("Enter a number to get factorial :"))
for i in range (num,1,-1):
    num=num*(i-1)
print("Factorial="+ str(num))

```

```python
# A program to find a number is whether prime or composite
num= int(input("Enter a number to be checked :"))
count=0
for i in range(num,0,-1):
    if num%i==0:
        count=count+1
if count==2:
    print("Prime number")
else:
    print("Composite number")
```

```python
#A program to check whether given string is palindrome or not
name= str(input("Enter a string"))
rev_name= reversed(name)
if list(name)== list(rev_name):
    print("A palindrome")
else:
    print("not a palindrome")
```

```python
#program to get the third side of right-angled triangle from two given sides
a=int(input("Enter first side value"))
b=int(input("Enter second side value"))
c=((a*a)+(b*b))**0.5
print(c)
```
